```python
import requests
import json

api_url = "http://omadev.cs.ucl.ac.uk/api"

protein = requests.get(api_url+"/protein/ANAPL04438/")


protein = protein.json()
print("The sequence of the protein is: \n")
print(protein['sequence'])

print("\nIt belongs the following oma group: \n ")
print(protein['oma_group'])

print("\nIt also belongs the following HOG: \n ")
print(protein['cdna'])
```
